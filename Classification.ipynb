```python
#Evaluation metrics for Classification---->we have Confusion Matrix
"""
In Confusion MAtrix--->we keep Actual and predicted in tabular form

                 Predicted
                 1              0
          1 | True +ve | False -ve
Actual      |
          0 | False +ve| True -ve

    1) Actual value is 1 and predicted value is 1--->we call it as-------->True Positive
     Here Employee is churn(1) and model predicted also as churn(1)------->True Positive

   2) Actual value is 0 and predicted value is 0--->we call it as---     --->True Negative
     Here Employee is not churn(0) and model predicted also as not churn(0)->True Negative

   3) Actual value is 0 and predicted value is 1--->we call it as------>False Positive
     Here Employee is not churn(0) but model predicted as churn(1)----->False Positive

   4) Actual value is 1 and predicted value is 0--->we call it as------>False Negative
      Here Employee is churn(1) but model predicted as not churn(0)----->False Negative


   In our predicted results-->If you have more True+ve's and True-ve's-->then it is Good
   If you have more False+ve's and False-ve's-->then it is Bad(ur model should not predict these)

   Your model should reduce False+ve's and False-ve's
"""
```

```python
#How Accuracy is calculated
"""
1)Accuracy=(TP+TN)/(TP+TN+FP+FN)

i.e Accuracy = ratio of correct predictions/total predictions

=========================================================================================
I want to focus to reduce False -ve's
ex: healthcare usecases
    patient has----> Covid
    but model predicted as ----->Non-Covid
    but covid case is a critical case

ex:2
    Banking customer----->is a defaulter(not paying the loans)
    but model predicted as non-defaulter
    It is a critical case

    For these kind of critical usecases like banking ,healthcare
    people concentrate more on to reduce False -ve's
    They want to avoid false -ve

    so apart from accuracy ,a new metric called  as--->Recall is used
    Recall also called as Sensitivity

    Recall=TP/(TP+FN)

    Recall ---->concentraing on False -ve's
=========================================================================================
Some usecases ----->more towards False +ve's

for that we have another metric called as ---->Precision

precision=TP/(TP+FP)

precision ---->concentraing on False +ve's
==========================================================================================
Some usecases focuses on reducing False -ves
some usecases focuses on reducing False +ves
Too many False+ve's and too many False-ve's are not good

so overall we want a balanced precision,

for that we have another metrics called as F1_score

F1_score=2*Precision*Recall/(Precision+Recall)

i.e F1_score is harmonic mean of Precision and Recall

whever people want balanced precision and recall ,they prefer F1_score

To test your model accuracy---->we go for F1_score

Based on Confusion matrix,

you can calculate your----------->accuracy
you can calculate your----------->precision(for False+ve's)
you can calculate your----------->recall(for False-ve's)
you can calculate your----------->F1_score(for both False+ve's and False-ve's balance)
                                            ie Precision+recall mean

all these we wont calculate manually

These metrics are specific to classification problems

Various possibilities of confusion matrix

   Actual      Predicted
     1            0   ----------->False -ve
     0            1   ----------->False +ve
     1            1   ----------->True +ve
     0            0   ----------->True -ve

These are given by probabilities

If you have too many TP & TN--->then ur accuracy will be high

If you have too many FP & FN--->then ur accuracy will be low

To overcome this,we have another metrics called as--->ROC-AUC score
==========================================================================================
ROC-AUC Score:

ROC--->Receiver Operating Characteristics
AUC--->Area Under Curve

As part of Accuracy,
they came up with 2 things
1)TPR(True Positive Rate) ----->as y-axis
2)FPR(False Positive Rate)----->as x-axis
TPR
|
|   /------------- whenever it is making a curve like this
|  /               we want to make sure, the area under curve(AUC) is more
| /            --->here the area is more and should be more,
|/_______________  that means your model is doing good
      FPR
TPR
|
|   /-------------
|__/_______________---->here the Area under the curve(AUC) is less
      FPR               that means your model is doing bad

whenever it is making a curve,make sure the area under curve is more--->model is doing good


TPR----->is nothing but Recall (False -ve)
         Recall=TP/(TP+FN)

FPR----->is nothing but False Positive Rate
         FPR=FP/(FP+TN)

AUC ---->is the ratio of TPR/FPR

AUC should be high----->then ur model is doing good
AUC should be low----->then ur model is doing bad

These all are the different metrics
which are related to classification

While implementing these--->we check for--->
            1.what is the accuracy
            1.what is Recall score
            2.what is precision score
            3.what is F1 Score
            4.what is ROC - AUC score
"""
```

```python
# #Decision Tree
# #Constructing a Tree like structure
# #It uses divide and conquer method
# #A Decision Tree is a Tree-structure,where each internal node represents a test
# #Each branch represents outcome of a test
# #Each Leafnode represnts class or Category
# #examples--->refer the diagrams

# #ex: A Customer approached for a loan
# """
#   i/p variable     o/pvariable
#   sal   age        Defaulter/Non-Defaulter (if non-defaulter--->u can approve the loan)
#    -     -          Yes
#    -     -          No
#    -     -          No

#    It builds a decision Tree

# It takes any one i/p variable(sal/age) -->where impurities are less

# here interanlly mthematical calculations

# here we have a method called------>Gini Impurity
# It is given as ======> Gini Impurity =1-sum(pi**2)
#                        here pi is ----->Probability

# Your Dataset should be pure
# For which i/p variable The Gini Impurity is less then that i/p variable
# will be taken as root node
# """
# """
# ex:        sal>40000
#              /  \ -->Here everytime,if sal>40000->directly we can label him as non-defaulter
#         yes /    \No Here there is no other conflict with defaulter
#            /      \  This kind of cases,we call it as homogeneous(only one class/category)
#   Defaulters=0    Defaulters=9
#   No-defaulters=18 Non-Defaulters=9 --->This is not homogeneous(bcoz of multiple categories)
#     |                 |                  IT is heterogeneous(2 classes/categories)
# (Homogeneous)         |                  (i.e 9 defaulters and 9 non-defaulters)
#                       |                  For heterogeneous we need to go for further splitting
#                     Age >30
#                      / \
#                 yes /   \No
#                    /     \
#     homogeneous  D=9     D=0   (homogeneous-->stop leaf node splitting or processing)
# stop splitting  ND=0     ND=9

# Whenever your leaf node reaches homogeneous then you can stop splitting the leaf node further

# This is how Decision Tree internally building tree structure for your dataset
# through which it is going to learn
# -here we used 2 columns(sal and age) but in real-usecases,it uses any no of columns

# -It uses a mechanism that which column can be put as root node and which column as
#  internal node and leaf node is calculated

# -Here one of the method to perform this is--->Gini Impurity =(1-sum(pi)square)

# -another method is---->Entropy = sum(-pi*log(pi))
#                                  i.e sum of negative probability score * log(pi)

# -Here we can go with either Gini Impurity or Entropy

#  we will be passing these (Gini Impurity /Entropy) as hyper parameters while
#  initializing the model

#  Here the Goal/Objective is --->which column can be put as root node or
#  which column if i select as rootnode so that u can traverse/reach the leaf node quickly

#  ex:  if i select salary, then i am able to reach the leaf node quickly

#  in which feature ur Gini Impurity/Entropy is less i.e pure and more pure
# i.e which is more homogeneous

#  if it is impure means, it is completely heterogenous

#  what we are achieving here --->whether we are able to traverse(reach) to leafnode quickly

#  -This is how Decision Tree Algorithm uses uses Gini impurity/Entropy in
#   computing/selecting rootnodes/internal nodes and leaf nodes through which it is
#   going to learn and predict

#   -Internally it builds complex nested if-else logic
```

**Random-Forest Classifier**

```python

# Random Forest ---->we call as Ensemble(Collections)

# collection of Decision Trees---->For classification

# also known as------>Bagging Algorithm

# #Here how the process is happening

# -You want to take a decision

# -Instead of taking decision from a single resource

# -It takes decisions from multiple resources

# -here the final decision is the best decision bcoz

#  u got advices from multiple resources through which you are learning and
#  consolidating which decision is the best one

#  -Normally, taking decisions from multiple resources will increase the confidence and
#   also will be more accurate

# #what Random Forest does is

# we build a Decision-Tree--->based on its traversing--->we get the leaf node
# ex:
# assume u have 100 records

# 1st random-Forest splits ur dataset into multiple sub-sets(samples)-->randomly it splits

# As a Datascientist,I will say--->I want Random Forest with 5 decision Trees
# or
# I want Random Forest with 5 Estimators(5 Decision Trees)

# whether to go with 5 Estimators(or) 50 Estimators--->we need to decide(explore)
# check the results for both and decide
# #ex:
# I want to build Random-Forest with 5 Estimators(5 Decision Trees)
# so it splits 100 recs into 5 Random-splits or 5Bags

# Bag is a collection of records


#     |----------100Records-----------------|
#     |       |        |      |             |---->Random Split
#    Bag1    Bag2    Bag3    Bag4        Bag5
#    (20)    (20)    (20)    (20)        (20)
#     |       |        |      |            |
#    DT1     DT2      DT3    DT4          DT5

#    It created multiple bags and for each bag---->a Decision Tree

#    and all are processed parallely
#    ex:
#    Forest----->collection of Trees
#    ex: u are passing 1 record to predict -->and check whether this candidate win/loose the
#    election

#    when you apply Random Forest classifier, It takes suggestions from all the Decision trees

#    Now for a given candiadate---->

#    DT1     DT2    DT3    DT4    DT5
#     |       |      |      |      |
#    Win     win    Loose  Win   Loose

#    Now Random Forest takes all the o/p's and check with the majority of votes

#    Majotity of votes---->win[3]
#                         Loose[2]
#   so the majority of votes---->is 3---->win

#   So finally the Random Forest predicts as---->win

#   all the remaining Decision Tree building is same as seen earlier
```

```python
# #why we call it as ensemble??

# Instead of getting processed with one single decision tree

# It is getting processed with multiple decision trees

# and ur Random Forest internally comprises of multiple models,
# that's why we call it as Ensemble(collection of trees)
```

```python
#Summary:
#i/p Records---->splitted into bag ofrecords
#for each bag,it generates a DecisionTree
#whenever you are prediting a record
#it gets o/p from each model
#and based on majority of votes,it predicts the majority class

#Normally in Datascience,we never say a particular model does well
#bcoz Datascience algorithms will fine tune based on ur datasets

#If your datasets are not good
#If your datasets are not quantitative
#If youe datasets are not qualitative
#ur algorithm wont do good

#so you need to explore with diff algorithms
#Some times Decision Tree does well
#sometimes RandomForest do well
#so explore and find the best one
```

**Naive Baye's Classifier**

```python

# Every Algorithn is following its own different way of learning the dataset through
# which it is going to PredictionErrorDisplay

# Logistic Regression--->It is following odds ratio and using maximum likelyhood estimation
#                        it is going to build the best sigmoid curve

# Decision Tree----->?it is internally building tree-structure through which it learns

# Random Forest----->It is building multiple decision trees ans using ensemble mechanism
#                   it is going to learn


# Naive Bayes
#     - Based on Baye's Theorem

#     -It is a conditional probability

#what's Baye's Theorem

# It is a pure probability based approach

# It is given as--->P(A/B)=(P(B/A)*P(A))/P(B)

# Ex: Probability of email is spam/not-spam

# P(A/B)=(P(B/A)*P(A))/P(B)

# here P(A/B)---->Posterior Probability
#      P(B/A)---->likelyhood Probability (probability of features given class)
#      P(A)---->Prior Probability
#      P(B)---->Evidence Probability(probaility of i/p)


# P(A/B)--->here A --->probability of class A(ex: spam) with given features B
#                B---->based on certain condition (one condition/2conditions/multiple conditions)
#                                                  (1 feature/2features/multiple features)
# It calculates the probability for Every feature seperately
# Finally it gets the consolidated result

# Naive Baye's for doing this calculations ----->there is an assumption that
# all i/p variables are independent-->there should not be any dependency

# that's why it is called naive

# Why it's Naive

# Naive Bayes assumes---->Each feature contributes independenlty to the final outcome

# How Naive Bayes works:

# 1.Calculate the prior probaility P(A) for each class --->P(A/B)

# 2.Calculate the likelihood(P(B/A)) for each feature per class

# 3.multiply all probabilities using Baye's formula

# 4.Pick the class with maximum probaility




# Ex: Spam Detection

# Features : words in an email

# ouput class :  spam/Not spam


# email contains(features such as)----->"free" ,"offer"


# Naive Bayes Computes
# P(Spam/free,offer)

# P(NoSpam/free,offer)

# whichever is higher probability---predict that

# ex: classify email--->as spam or Not spam ---->based on 2 words
#     1)Free
#     2)Offer

#     #P(Spam/Free,Offer)=P(Free/Spam)*P(Offer/Spam)*P(Spam)
#     #P(notspam/Free,offer)=P(Free/notspam)*P(offer/notspam)*P(notspam)

#     Training Data

#     Email    Free  offer   class
#      1        yes    yes    spam
#      2        yes    No     spam
#      3        No     yes    Not spam
#      4        No     No     Not spam

#      Total emails=4

    #  Step 1: Calculate prior probabilities P(A)

    #     P(Spam)=2/4=0.5

    #     P(Not Spam)=2/4=0.5



    #   Step 2: Calculate the Likelihood probabilities (p(B/A))
    #             for each feature per class
    #           i.e P(Free/Spam)
              # i.e P(Offer/Spam)

              #  i.e P(Free/Not Spam)
              #  i.e P(Offer/Not Spam)
    #    i)For spam
    #    P(Free = Yes/spam) =2/2=1
    #    P(offer=Yes/spam) =1/2=0.5

    #    ii)For Not spam
    #    P(Free = Yes/Not spam) =0/2 =0
    #    P(offer=Yes/Not spam) =1/2=0.5


    #Step 3: classify a new Email
    #        Assume Free=yes and offer=yes

    #     Apply Baye's theorem -->compute posterior probability
    #      for spam
    #      P(Spam/Free,Offer)=P(Free/Spam)*P(Offer/Spam)*P(Spam)
    #                         =  1  *0.5 *0.5
    #                         =0.25
    #      for Not spam
    #      P(Not Spam/Free,Offer)=P(Free/Not Spam)*P(Offer/Not Spam)*P(NotSpam)
    #                        =  0  *0.5 *0.5
    #                        = 0

    #      Final prediction
    #        0.25 >0 ---->Email is classified as spam

 #==================================================================================
 #Bayes Theorem:

 #P(A/B)=(P(B/A)*P(A))/P(B)

 #P(Class/Features)=(P(Features/class)*P(class))/P(Features)

 #P(class/Features) i.e P(A/B)--->posterior probability (what we want)

 #P(B/A) i.e P(Features/Class)--->Likelyhood

 #P(class)--->Prior probability

 #P(Features)->Evidence(same for all classes)--->ignore while comparision

 #How Naive Bayes works or classifies

 #1.Compute the prior probabilties for each class  p(A)

 #2.Compute likelyhood of each feature in the class p(B/A)

 #3.multiply the above two

 #4.Pick the class with the highest probability

 #============================================================================
 #Ex:2
 #    Email Text               class
 #    "win money-2Lakhs"       Spam
 #    "Limited win project"    spam
 #    "DS project at 9pm"      not spam
 #    "No project class today" not spam

 #Step 1: compute Prior probabilities
 #        P(spam)=2/4=0.5    P(NotSpam)=2/4=0.as_integer_ratio

 #Classify--->"win","project"
 #Step 2: compute Likelihoods
 #        P("win"/spam)=2/2=1
 #        P("win"/notspam)=0/2=0

 #        P("project"/spam)=1/2=0.5
 #        P("project"/notspam)=2/2=1

 #Step 3:

   #Apply Baye's theorem
    #  comute posterior probability
    #      for spam
    #      P(Spam/win,project)=P(win/Spam)*P(project/Spam)*P(Spam)
    #                         =  1 *0.5*0.5
    #                         = 0.25
    #      for Not spam
    #      P(Not Spam/win,project)=P(win/Not Spam)*P(project/Not Spam)*P(NotSpam)
    #                        =  0  *1 *0.5
    #                        = 0

    #      Final prediction
    #        0.25 >0 ---->Email is classified as spam

  #===========================================================================
 #Ex:3

  #  Weather          Temperature        play tennis
  #  Sunny               Hot               No
  #  Sunny               mild              No
  #  overcast            cool              yes
  #  Rainy               Mild              yes

  #Task: will there will be a play when weather=Sunny Temperature=cool ???

   #Step 1: compute Prior probabilities
 #        P(yes)=2/4=0.5    P(No)=2/4=0.5

 #Classify--->"Sunny" ,"cool"
 #Step 2: compute Likelihoods
 #        P("sunny"/yes)=0/2 =0
 #        P("sunny"/no)=2/2=1

 #        P("cool"/yes)=1/2=0.5
 #        P("cool"/no)=0/2=0


#Step 3:

   #Apply Baye's theorem
    #  comute posterior probability
    #      for class "yes"
    #      P(yes/sunny,cool)=P(sunny/yes)*P(cool/yes)*P(yes)
    #                         =  0*0.5*0.5
    #                         = 0
    #      for class "No"
    #      P(No/sunny,cool)=P(sunny/No)*P(cool/No)*P(No)
    #                        =  1  *0 *0.5
    #                        = 0



    #      Final prediction--->internally it applies smoothing
    #
    #==========================================================================
  #ex:4 Medical Diagnosis (Binary Features)

  #Fever        Cough        Flu
  # yes          Yes         yes
  # yes          No          yes
  #No            Yes         No
  # No           No          No

  #predict for a new patient - Fever=yes cough=yes

  #Step 1: Prior probability
   #        P(A)--->probability of each class
   #        P(Flu)=2/4=0.5
   #        P(NoFlu)=2/4=0.5

  #step 2: likelihoods: P(B/A)
         #For the new patient
  #   P(Fever=Yes/Flu)=2/2=1
  #   P(cough=yes/Flu)=1/2=0.5

  #for noflu
  #  P(Fever=Yes/noFlu)=0/2=0
  #  P(Cough=yes/NoFlu)=1/2=0.5

  #Step 3:posterior probability
  #       P(flu/fever,cough)=P(fever/flu)*P(cough/flu)*P(flu)
  #                         =1*0.5*0.5
  #                         =0.25

  #       P(Noflu/fever,cough)=P(fever/noflu)*P(cough/noflu)*P(noflu)
  #                           =    0  * 0.5 *0.5
  #                           =0

  #Select the class with high probability--->P(flu) > P(Noflu)
  #                                           0.25  > 0

  #Conclusion :Flu predicted

#======================================================================================

#Types of Naive Baye's Classifiers:
#1)Gaussian Naive Baye's----->when features are continous
#                            It calculates the mean and standard deviation for each class
#                            to predict the outcomes
#                            suitable for data like-->height,weight,sensor measurements

#2.Multinominal "   " ------>For text classification where features represents wordcounts
#                            or frequencies---->mostly for frequency data
#                            ex: spam detection

#3.Bernoulli    "   " -----> For binary or boolean features(i.e 0/1,present/absent,yes/No)
#
#4.Categorical  "   "------>for categorical  data

#Multinomial and categorical are almost same

#which one is the best---->Gaussian/Bernoulli/multinominal/Categorical

#We cant say---->we need to explore---->as a hyper parameter(learning paramter)

#which variant is doing good,we need to explore
```

**K-Nearest Neighbours(KNN) Classifer**

```python
"""
KNN is  supervised machime LEarning algorithm used mainly for classification

It classifies a data point based on how its nearest neighbours are classified

Core methodology---->Distance metrics
                     based on closest distance, it learns/predicts

KNN says--->"Tell me who your neighbours are,I'will tell you who you are"

KNN is ----->"you are like your neighbours" algorithm

KNN is also called as a Lazy learner
bcoz, it doesnt perform anything during training phase but it starts
 computing at prediction time

 during predictiontime--->it applies the distance metrics

KNN Process:
  -when a new datapoint is given
  -It looks at the k closest and nearest datapoints in the training set
  -among those neighbours the class which has majority will be considered


HOw KNN works:

1.choose the value of k
  k=number of nearest neighbours to consider(ex,k=3 or k=5 or k=7)
  if k=5 ,then u are checking with 5 nearest neighbours

2.Calculate the distance
  measure the distance b/w the new datapoint and all other training points
  for calculating distance, we have 2 distance metrics
  i)Euclidean distance(mostly used)
 ii)Manhattan distance

3.Find k nearest neighbours
  select k-datapoints with  the smallest/nearest distance

4.Vote for the class
  The class with the maximum votes, that class is assigned to the new datapoint

  ex: k=5 and among 5 nearesst neighbours
      3 belong to classA
      2 belong to ClassB

      majority=3 belongs to class A

      So the new datapoint is classified as---> class A
"""
```

```python
"""
we also call KNN as a Lazy learner
bcoz, it doesnt perform anything during training phase but it starts
 computing at prediction time

Applications:
  -pattern recognition
  -Image classifcation
  -Medical diagnosis
  -credit scoring
"""
"""
ex:1

   point        X         Y        class
    A           1         2          Red
    B           2         3          Red
    C           3         3          Blue
    D           6         5          Blue
    E           7         6          Blue

  New point to classifiy---> p=(2.5,2.5)

  #Step 1: It Compute distances

  Distance to A------>near
  Distance to B------>Very near
  Distance to C------>far
  Distance to D------>too far
  Distance to E------>too fat

  #step 2: Choose the value of k
       say k=3

   Nearest neighbours are:
      A(Red)
      B(Red)
      C(Blue)

  #step 3: MAjority vote
      Red-2
      Blue-1

  Prediction :Red


  commonly used Distance metrics is----->Euclidean Distance

  for two points p=(x1,x2)  q=(y1,y2)

  Distance =sqrt((x1-y1)square +(x2-y2)square)


EX: churn/Not-churn
             * * *
    |       * * *  ----->non-churn
    |        * * *
    | **   x ----->new given point (for this it checks the k-nearest points)
    |* * *------>churn
    | * **
    |___________________

    KNN ia a lazy learner--->During training phase it just plots
    And during prediction time it applies the distance metrics and gets the k-nearest
    neighbours and takes majority from that to predict the class

 ex:
     |
     |         * (4,3)
     |
     |          --------->Computing the distance b/w 2 points
     |  *(2,1)            Euclidean distance =sqrt((4-2)square+(3-1)square)
     |____________________

2)another distance metrics is----->Manhattan distance
   Distance =|x1-y1|+|x2-y2| (here istead of taking sqt ans square, it takes absolute value)

 so Distance=|2-4|+|1-3|

 default distance metrics used by KNN is-----> Euclidean distance

 where exactly  and when it uses distance metrics

 during predicting--->if any new record is kept to predict



 ex: k- here k means, how many neighbours,you want to validate
     ex: 5
     it checks for 5 neighbours
     in that 5 neighbours---> Non-churn =2
                              churn=3
                so finally predict the majority class---->say --->churn

-------------------------------------------------------------------------------
General equation for distance metrics
  Sum[(O1i-O2i)**p]**1/p

we call this as minkowski distance

here based on p value----->it decides--->whether it is Euclidean or Manhattan

p=1--->Manhattan distance
p=2--->Euclidean distance
in the above formula---->if you substiute p=2 then we get Euclidean distance formula
                    ----> "        "      p=1 then we get Manhattan distance formula
"""
```
