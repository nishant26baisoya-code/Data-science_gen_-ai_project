# SETS

```python
#Sets : collection of elements

#Set properties

#1.Sets represented using {}

#2.Sets allows heterogenous elements

#3.Sets doesnt allow duplicates x={10,20,30,10,30}

#4.Sets are un-ordered x={10,20,30,40,50}

#5.Sets doesnt support indexing

#6.Sets are mutable ----> changes and modifications are allowed

#7.sets allows only immutable,sets wont allow mutables, even set not allowed within a set
```

```python
x={10,20,30,40,50}
print(x,type(x))

#set methods --> add(),update(),pop(),remove(),clear()
#1.add() :To add an element in to the set
x.add(60)
print(x,len(x))

x.add((55,65,75)) #tuples gets added
print(x,len(x))

#2.update(): for updating multiple elements ata time
x.update((70,80,90)) #here 3 elements gets added
print(x,len(x))

#3.pop() for removing a radomly
x.pop()
print(x)
x.pop()
print(x)

#4.remove() :For removing a particular element
x.remove(40)
print(x)
x.remove((55,65,75))
print(x)

#5.clear() :clears all the set elements
x.clear()
print(x)

#set mathematicals methods
A={10,20,30,40,50,60}
B={10,20,30,70,80,90}

#6.union():union of 2 sets
print(A.union(B))

#7.intersection() : returns common elements of
print(A.intersection(B))

#8.difference():
print(A.difference(B))
print(A-B)
print(B.difference(A))
print((B-A))
```

# Dictionary

```python
#Dictionary :Dictionary is a collection (key,value) pairs
ex: x={"name","Ajay","age":25,"city":"pune"}
         k      v     k    v    k      v
Dictionary properties:

1.No insertion order
2.No duplicate keys but duplicate value allowed
  ex: x="age":60,"weight":60}

3.No indexing

4.Dictionary is a mutable object--->changes and modication allowed
                                    i.e we can insert, remove and modify
5.If we pass the key we get the value
```

```python
x={}
print(type(x))

x={'a':10,'b':20}
print(x,type(x),len(x))

#If we pass the key,we get the value
print(x['a'])
print(x['b'])

#duplicates keys are not allowed
x={'a':10,'b':20,'b':30,'a':40}
print(x,len(x))
print(x['a'])
print(x['b'])

#modifying adding and removing element from a dictionary
x={"name":"Ajay","age":25}
print(x)
#modify the age to 28
x["age"]=28
print(x)
#adding a new key--->city
x["city"]="chennai"
print(x)

#removing the age
x.pop("age")
print(x)

#clearing all the elements
x.clear()
print(x)
```

```python
#using lambda as an argument to other function like filter(),map()
#filter () : for filtering based on data
#filter takes 2 arguments
#1.lambda as 1st argument
#2.collection like list tuple

list1=[10,20,30,40,50,60,70]
#Task :filter only those whose value is above 40
res=list(filter(lambda x:x>40,list1))   #here x is element of the list,which is of type int
print(res)

#ex:2
emps=[[101,"ajay",10000,'m',11,'hyd'],
      [102,"rahul",20000,'m',12,'pune'],
      [103,"priya",300000,'f',13,'hyd'],
      [104,"Anitha",40000,'f',14,'hyd'],
      [105,"jame",500000,'m',15,'pune']
     ]
#Task1 : Filter records only "hyd" employee records
hyd_recs=list(filter(lambda x:x[5]=="hyd",emps))
print(hyd_recs)
print("\n")

#Task2 : Filter only male emp records
print(emps[0].index('m'))
male_recs=list(filter(lambda x:x[3]=='m',emps))
print(male_recs)
print("\n")

#Task3 : Filter those emps whose salaries above 30000 and should belong to dno 12
res2=list(filter(lambda x:x[2]>30000 and x[4]==12,emps))
print(res2)
```

```python
#map() :for each element transformation
#map() also takes 2 arguments
#1.lambda
#2.collection like list

#ex:1
list1=[10,20,30,40,50]
#task :adding 5 each of element of the list
res=list(map(lambda x:x+5,list1)) #here x is each element of list of type int
print(res)

#ex:2
emps=[[101,"ajay",10000,'m',11,'hyd'],
      [102,"rahul",20000,'m',12,'pune'],
      [103,"priya",300000,'f',13,'hyd'],
      [104,"Anitha",40000,'f',14,'hyd'],
      [105,"jame",500000,'m',15,'pune']
     ]
#Task1 : Adding 5000 to the sals of each employee as bonus
res2=list(map(lambda x:[x[0],x[1],x[2]+5000,x[3],x[4],x[5]],emps))
print(res2)

#Task2 : Transform each empname to start with uppercase character
res3=list(map(lambda x:[x[0],x[1].capitalize(),x[2]+5000,x[3],x[4],x[5]],emps))
print(res3)
```

```python
#Dataframe filtering operations

#Filtering data from a Dataframe

# importing pandas
import pandas as pd

data = {
  'Name': ['Miller','Blake','James','Jimmy','Sony','Ruby' ],
  'Age': [23,24,22,25,23,22],
  'Gender':['m','m','m','f','f','f'],
  'Branch':['CSE','ECE','EEE','IT','CSE','ECE'],
  'Percentage': [84,82,78,76,68,56],
  }
# Create a dataframe
df=pd.DataFrame(data, columns =['Name','Age','Gender','Branch','Percentage'])
print(df)
print("\n")

#-------------------------------------------------------------------------------------------------------

#i.filtering data based on percentage

df1 = df[df['Percentage'] > 80]
print(df1)
#Note: If we filter a Dataframe then the result is also a Dataframe
print("\n")

#or

df1 = df.loc[df['Percentage'] > 80]
print(df1)

print("\n")
#------------------------------------------------------------------------------------------------------

#ii. I want only the CSE students

df1 = df[df['Branch'] == 'CSE']
print(df1)

print("\n")
#------------------------------------------------------------------------------------------------------
#iii. I want other than CSE students
df1 = df[df['Branch'] != 'CSE']
print(df1)

print("\n")

#Note : we can use == , != , >= ,<= ,> , < for filtering the data

#-----------------------------------------------------------------------------------------------------
#iv : isin() method()

#ex:
#I require only those students who belongs to CSE or IT background

options=['CSE','IT']

df3 = df[df['Branch'].isin(options)]
print(df3)

print("\n")

#or
df4 = df.loc[df['Branch'].isin(options)]
print(df4)

print("\n")
#------------------------------------------------------------------------------------------------------

#V :I want to know those students who are from non-CSE background
options=['CSE','IT']

df5 = df[~df['Branch'].isin(options)]
print(df5)
print("\n")
#------------------------------------------------------------------------------------------------------

#Vi. Providing multiple conditions
#I want those students whose percentage < 70 and are non-CSE Background
df6 = df[(df['Percentage'] < 70) &
        ~df['Branch'].isin(options)]
print(df6)
```

```python
#Grouping and aggression in pandas
import pandas as pd
df = pd.read_csv('/content/orders.csv') #df is a DataFrame.
#print (df)

#Reading Specific Rows of a particular column--->5 rows
print (df[0:5]['customer'])

#Reading Specific column
'''We use the multi-axes indexing method called loc() for this purpose
We choose to display the customer and order column for all the rows.'''
#print (df.loc[:,['customer','order']])

#Reading Specific Columns and Rows
'''The read_csv function of the pandas library can also be used to read some specific columns and specific rows.
We use the multi-axes indexing method called .loc() for this purpose.
We choose to display the customer and order column for some of the rows.'''
print(df.loc[[1,3,5],['customer','order']])

#Reading Specific Columns for a Range of Rows
print (df.loc[2:6,['customer','order']])

#------------------------------------------------------------------------------

#Various Aggregations--->sum(),mean(),max(),min(),count()

#How many rows the dataset has i.e how many total orders were made--->using count()
#1.No of orders----->using count()
orders=df['order'].count()
print("NO OF ORDERS:",orders)

#2.Total order value(TOV) generated for all the orders ---->sum()
tot_rev=df['revenue'].sum()
print("TOTAL ORDER VALUE:",tot_rev)

#3.Average order value(AOV) generated for all orders ---->mean()
avg_rev=df['revenue'].mean()
print("AVERAGE ORDER VALUE(AOV):",avg_rev)

#4.Max order value  ---->max()
max_rev=df['revenue'].max()
print("MAX ORDER VALUE:",max_rev)

#5.Min order value  ---->min()
min_rev=df['revenue'].min()
print("MIN ORDER VALUE:",min_rev)

# TASK-1. Calculate total number of orders per customer;
orders_cust=df.groupby('customer')['order'].count()
print("Total No of orders per customer:\n",orders_cust)

#TASK-2: Calculate TOV per customer, TOV  - Total revenue;
TOV_cust=df.groupby('customer')['revenue'].sum()
print("TOV per Customer:\n",TOV_cust)

#TASK-3: Calculate AOV per customer, AOV  - average revenue;
AOV_cust=df.groupby('customer')['revenue'].mean()
print("AOV per Customer: \n",AOV_cust)

# Calculate the count of orders based on order_state
cnt=df.groupby('order_state')['order_state'].count()
print("order_state:\n",cnt)

#multi grouping
#df.groupby(['col1', 'col2']).agg({'col3': [min, max, sum]})
#o/p:
'''
citywise,dnowise---->totsal
hyd 11 --->totsal
hyd 12 --->totsal
hyd 13 --->totsal
pune 11 -->totalsal
pune 12 -->totsal
pune 13 -->totsal
'''
df2 = pd.read_csv('emp1.csv')
df3=df2.groupby(['city','dno']).agg({'Salary':[sum]})
print(df3)
#o/p:
'''
city dno
city dno
hyd  11   10000
     12   30000
     13   50000
pune 11   40000
     12   20000
'''
#-------------------------------------------------------------------------------

#Filtering data:
#Selecting rows in pandas DataFrame based on conditions

#I want only those orders whose ordervalue(revenue) >100
df1 = df[df['revenue'] > 100]
print(df1)
print("\n")
#or
df2 = df.loc[df['revenue'] > 100]
print(df2)
print("\n\n\n")

print("\n")

#I want those orders which are not completed
#ex:2
df3 = df.loc[df['order_state']!='completed']
print(df3)
```
