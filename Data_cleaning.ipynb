```python
#Replacing using mean()
import pandas as pd
details=[['Ajay',501,90,80,70,18],
         ['Rahul',502,60,40,30,],
         ['Amar',503,30,90,60,18],
         ['Zaheer',504,190,80,70,16],
         ['Shiva',505,55,75,80,],
         ['Amar',503,30,90,60,18],
         ['Zaheer',504,190,80,70,16],
        ]
#Creating a dataframe
df=pd.DataFrame(details,columns=["NAME","ROLLNO","MATHS","PHYSICS","CHEMISTRY","AGE"])
print(df)
```

```python
#Now Replacing with mean()
mean=round(df['AGE'].mean())
print("mean=",round(mean))
df.fillna({'AGE':mean},inplace=True)
print(df)
```

```python
#Replace the null values using median
median=df['AGE'].median()
print("median=",median)
df.fillna({'AGE':median},inplace=True)
print(df)
```

```python
#Replacing using mean()
import pandas as pd
details=[['Ajay',501,90,80,70,18],
         ['Rahul',502,60,40,30,],
         ['Amar',503,30,90,60,18],
         ['Zaheer',504,190,80,70,16],
         ['Shiva',505,55,75,80,],
         ['Amar',503,30,90,60,18],
         ['Zaheer',504,190,80,70,16],
        ]
#Creating a dataframe
df=pd.DataFrame(details,columns=["NAME","ROLLNO","MATHS","PHYSICS","CHEMISTRY","AGE"])
print(df)
```

```python
#Replacing wrong data
df.loc[[3,6],'MATHS']=90
print(df)
```

```python
#If marks entered above >100 then replace then with 0
for p in df.index:
  if(df.loc[p,"MATHS"]>100):
    df.loc[p,"MATHS"]=0

print(df)

```

```python
print(df) #excuting the previous cell)
```

```python
#Another way--->if you have got huge data--> only only 4-5 rows having
#Wrong data (marks>100) then remove those rows which wont affect ur result

for p in df.index:
  if(df.loc[p,"MATHS"]>100):
    df.drop(p, inplace=True)

print(df)
```

```python
#Removing the duplicates

df.drop_duplicates(inplace=True)
print(df)
```

```python
#Removing the unnecessary data(columns)
#dropping the columns which doesnt have importance()
df1=df.drop(columns=["ROLLNO","AGE"])
print(df1)
```
