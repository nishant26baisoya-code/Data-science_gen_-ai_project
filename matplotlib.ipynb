```python
#Comparing the scores
# INDIA VS PAK
#Overs as x-axis
#runs_india,runs_pak as y-cordinates

import matplotlib.pyplot as plt

overs=[10,20,30,40,50]

runs_India=[75,110,195,230,290]

runs_pak=[50,120,155,205,260]

plt.plot(overs,runs_India,color="blue",marker="o",markersize=12,label="INDIA")

plt.plot(overs,runs_pak,color="green",linestyle="dashed",linewidth="2",label="PAKISTAN")

plt.xlabel("OVERS")

plt.ylabel("RUNS")

plt.title("SCORE COMPARISIONS")

plt.legend()

plt.show()
```

```python
#pie chart:
import matplotlib.pyplot as plt

hours = [7,2,2,13]

activities = ['sleeping','eating','working','playing']

cols = ['g','y','r','b']

plt.pie(hours,
        labels=activities,
        colors=cols,
        startangle=90,        #starting from 90 angle
        shadow=True,
        explode=(0,0,0.3,0),  # for highlighting particular activity
        autopct='%1.2f%%'     # for showing the percentage
        )

plt.title('Interesting Graph\ncheck it out')
plt.show()
```

```python
#stackplot

import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('days')
plt.ylabel('hours')
plt.title('Interesting Graph\ncheck it out')
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating  =  [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)
plt.plot([],[],color='k',label='Playing',linewidth=5)

#empty list bcoz plot() expecting 2 parameters to plot ,so given empty

plt.stackplot(days,sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('days')
plt.ylabel('hours')
plt.title('Intersting Graph\ncheck it out')
plt.legend()
plt.show()
```
