```python
#import the required Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
#Loading the dataset
df=pd.read_csv("loan_eligibility_dataset.csv")

#Display first few rows to understand the dataset
print(df.head())
```

```python
#Selecting numerical column for scaling
numerical_features=["Income","LoanAmount","CreditScore","DebtToIncomeRatio"]

#Crearting a dataframe for numerical features
df_numerical=df[numerical_features]


```

```python
#Initilaize Standardscaler
standard_scaler=StandardScaler()

#Apply standard scaler Transformation
df_standard_scaling=standard_scaler.fit_transform(df_numerical)
print(df_standard_scaling)

#Converting to a Dataframe
df_standard_scaled=pd.DataFrame(df_standard_scaling,columns=numerical_features)

#Display the scaled data
print("\n Standard Scaled Data(Z-score normalization):\n",df_standard_scaled.head())
```

```python
#MinMaxScaler Method
```

```python
#initialize MinMaxScaler
min_max_scaler=MinMaxScaler()

#Apply MinMaxScaler Transformation(on the i/p features dataframe)
minmax_scaling=min_max_scaler.fit_transform(df_numerical)

#converting/creating a DataFrame
df_minmax_scaled=pd.DataFrame(minmax_scaling,columns=numerical_features)

#Display the scaled data
print("\n Min-Max Scaled Data (Range 0-1):\n",df_minmax_scaled.head())
```

```python
#Save standard scaled Dataset
df_standard_scaled.to_csv("standard_scaled_loan_data.csv",index=False)
#Save Min-Max Scaled DataSet
df_minmax_scaled.to_csv("minmax_scaled_loan_data.csv",index=False)
```
