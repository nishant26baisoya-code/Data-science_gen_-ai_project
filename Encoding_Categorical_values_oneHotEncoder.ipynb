```python
import pandas as pd
#Load the dataset create a dataframe
df=pd.read_csv("customer_segmentation.csv")
```

```python
#Display first few rows to inspects the datasets
print(df.head())
```

```python
#printing the datatypes of each elements
print(df.dtypes)
```

```python
#Define the categorical columns that requires encoding
categorical_cols=["Region","ProductCategory","MembershipType"]
```

```python
#Here we use scikit-learn library --->pre-installed in colab
#To download and install sklearn---->using pip
#!pip install sckit-learn
import sklearn   #for importing sklearn components
print(sklearn.__version__)

#if any issues while installing
#u can upgrade while installing saying---> !pip install -U sckit-learn

#ex: for importing sklearn components
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(drop="first",handle_unknown='ignore')
```

```python
#Apply encoding on categorical columns-->Applying fit and Transform
encoded_data=encoder.fit_transform(df[categorical_cols]).toarray()
#It generates new columns
```

```python
#convert encoded data to dataframe with meaningfull column names
#we are creating a dataframe from a encoded_Data
encoded_df=pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(categorical_cols))
print(encoded_df.head())
```

```python
encoder.get_feature_names_out(categorical_cols)
```

```python
#here we see only the encoded columns
#Now combine encoded columns with the original dataset(excluding the original categorical columns)
df_encoded=pd.concat([df.drop(columns=categorical_cols),encoded_df],axis=1)

#Now display the Transformed dataset
print("\n Dataset after one-hot Encoding:\n",df_encoded.head())
```
